```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import matplotlib.pyplot as plt

# --- 1. HİPERPARAMETRELER VE AYARLAR ---
# Bilimsel Not: Deneyin tekrarlanabilirliği için parametreleri sabitliyoruz.
BATCH_SIZE = 32
IMG_SHAPE = (150, 150) # Görüntü boyutu
EPOCHS = 10 # Modeli kaç tur eğiteceğiz? (Başlangıç için 10 ideal)

# Dosya yolları
base_dir = '../data/raw'
train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

# --- 2. VERİ ÖN İŞLEME VE ARTIRMA (DATA AUGMENTATION) ---
# Bilimsel Not: Eğitim verisini çeşitlendirerek (döndürme, kaydırma) 
# modelin ezberlemesini (overfitting) engelliyoruz.
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,      # Rastgele 40 derece döndür
    width_shift_range=0.2,  # Sağa sola kaydır
    height_shift_range=0.2, # Yukarı aşağı kaydır
    shear_range=0.2,        # Bükme
    zoom_range=0.2,         # Yakınlaştırma
    horizontal_flip=True,   # Yatay çevirme
    fill_mode='nearest'
)

# Test verisine ASLA değiştirme (augmentation) yapmıyoruz, sadece normalize ediyoruz.
# Çünkü gerçek dünya verisi gibi saf kalmalı.
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SHAPE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SHAPE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

# --- 3. CNN MODEL MİMARİSİ TASARIMI ---
# Bilimsel Yöntem: Özellik çıkarımı için Konvolüsyon katmanları kullanıyoruz.
model = Sequential([
    # 1. Konvolüsyon Bloğu: Temel özellikleri (kenar, çizgi) yakalar
    Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(2, 2),

    # 2. Konvolüsyon Bloğu: Daha karmaşık şekilleri yakalar
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    # 3. Konvolüsyon Bloğu: Detayları ve dokuları (çürükleri) yakalar
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    
    # 4. Konvolüsyon Bloğu: Derinlemesine özellik analizi
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    # Sınıflandırma Katmanı (Karar verici kısım)
    Flatten(),
    Dropout(0.5), # Nöronların %50'sini kapatarak ezberlemeyi önler
    Dense(512, activation='relu'),
    Dense(6, activation='softmax') # 6 Sınıfımız olduğu için çıkış 6 nöron (Softmax: Olasılık verir)
])

# Modeli derleme
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Model özetini görelim (Raporuna bu tablonun çıktısını koyabilirsin)
model.summary()

# --- 4. EĞİTİMİ BAŞLATMA ---
print("\n--- Model Eğitimi Başlıyor ---")
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE, # Tüm veriyi bir turda gör
    epochs=EPOCHS,
    validation_data=test_generator,
    validation_steps=test_generator.samples // BATCH_SIZE
)

# --- 5. SONUÇLARI GÖRSELLEŞTİRME ---
# Başarım (Accuracy) Grafiği
plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'], label='Eğitim Başarısı (Train Acc)')
plt.plot(history.history['val_accuracy'], label='Doğrulama Başarısı (Val Acc)')
plt.title('Eğitim ve Doğrulama Başarısı')
plt.xlabel('Epoch (Tur Sayısı)')
plt.ylabel('Başarı Oranı')
plt.legend()
plt.show()

# --- 6. MODELİ KAYDETME ---
# Eğittiğimiz modeli kaydedelim ki tekrar tekrar eğitmeyelim.
if not os.path.exists('../models'):
    os.makedirs('../models')
model.save('../models/fruit_cnn_model.h5')
print("Model başarıyla kaydedildi: models/fruit_cnn_model.h5")
```
